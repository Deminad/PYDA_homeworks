Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

```python
import pandas as pd
```

```python
# Задание 1
ratings = pd.read_csv("/Users/Stanislav/Desktop/ml-25m/ratings.csv")
```

```python
def movie_class(rate):
    if rate <= 2:
        return "low" 
    elif rate <= 4:
        return "middle" 
    else: return "high"
```

```python
# Задание 2 
kw = pd.read_csv("keywords.csv")
```

```python
def geo_tag(keyword):
    geo_data = {
                'Центр': ['москва', 'тула', 'ярославль'],
                'Северо-Запад': ['петербург', 'псков', 'мурманск'],
                'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
                } 
    for region, cities in geo_data.items():
        for city in cities:
            if city in keyword:
                return region
            else: return "undefined"
        
```

```python
kw['region'] = kw['keyword'].apply(geo_tag)
kw.sort_values('region', ascending=False).head(10)
```

```python
# Задание 3
ratings = pd.read_csv("/Users/Stanislav/Desktop/ml-25m/ratings.csv")
movies = pd.read_csv('/Users/Stanislav/Desktop/ml-25m//movies.csv')
ratings = ratings.merge(movies, on='movieId', how='left')
years = [str(x) for x in range(1950,2011,1)]
```

```python
def production_year(title, years):
    for year in years:
        if year in title:
            return year
        
    return '1900'
```

```python
ratings['year'] = ratings['title'].apply(production_year, years = years)
(ratings
 .groupby('year')
     .mean()['rating']
     .sort_values(ascending=False)
 .head(10)
)
```
