Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv', sep = ',')
ratings.head()
```

```python
def calc_lifetime(list_):
    return round((max(list_) - min(list_))/86400, 2)
```

```python
#Задание 1
previous_user = None
time_list = []


with open('ml-latest-small/ratings.csv') as f:
    header = f.readline()
    
    for line in f:
        line = line.strip().split(',')
        
        user_id, movie_id, rating, timestamp = line
        timestamp = int(timestamp)
        
        if previous_user == None:
            previous_user = user_id
        else:
            if user_id == previous_user:
                time_list.append(timestamp)
            else:
                # вычисление
                if len(time_list) < 100: 
                
                    time_list = []
                    time_list.append(timestamp)
                    previous_user = user_id
                
                else:
                    calculation = calc_lifetime(time_list)
                    print(previous_user, calculation, len(time_list))
                
                    time_list = []
                    time_list.append(timestamp)
                    previous_user = user_id
        
calculation = calc_lifetime(time_list)
print(previous_user, calculation, len(time_list)) 
```
